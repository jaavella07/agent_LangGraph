```typescript
import { StateGraph, START, END } from "@langchain/langgraph";

// Definición del estado
type State = {
  customer_name?: string;
  my_age?: number;
};

// Nodo equivalente
function node_1(state: State): Partial<State> {
  if (state.customer_name == null) {
    return {
      customer_name: "John Doe",
    };
  }
  return {};
}

// Construcción del grafo
const builder = new StateGraph<State>();

builder.addNode("node_1", node_1);

builder.addEdge(START, "node_1");
builder.addEdge("node_1", END);

// Compilar agente
const agent = builder.compile();

// Dibujar el grafo 
//console.log(agent.getGraph().drawAscii());
```
