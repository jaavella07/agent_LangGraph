```typescript
import { initChatModel } from "langchain/chat_models/universal";
import { ChatOpenAI } from "@langchain/openai";

// Inicializar el modelo
const llm = await initChatModel("gpt-4o", {
  modelProvider: "openai",
  temperature: 0,
}) as ChatOpenAI;

// Definir el tool de file_search
const fileSearchTool = {
  type: "file_search" as const,
  vector_store_ids: ["vs_69cc756a9338819188f6164a44826de2"],
};

// Bindear el tool al modelo OPENAI
const llmWithTools = llm.bindTools([fileSearchTool]);

// Invocar el modelo
const response = await llmWithTools.invoke("¿Qué entendemos por Pensamiento Computacional?");

// obtener solo el texto plano
const text = response.text;
console.log("TEXT:", text);
```
