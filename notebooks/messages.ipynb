```typescript
import { AIMessage, HumanMessage } from "@langchain/core/messages";

// Crear mensajes
const aiMsg = new AIMessage({
  content: "Message from the AI",
});

console.log(aiMsg.content); // equivalente a .text en Python

const humanMsg = new HumanMessage({
  content: "Message from the Human",
});

console.log(humanMsg.content);

// Historial
const history = [aiMsg, humanMsg];

// Recorrer e imprimir
for (const m of history) {
  console.log(`[${m._getType()}]: ${m.content}`);
}

// Agregar un nuevo mensaje
const newHistory = [
  ...history,
  new AIMessage({ content: "Message from the AI" }),
];

console.log(newHistory);
```
